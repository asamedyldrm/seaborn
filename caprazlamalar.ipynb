### Çaprazlamalar

```python
# ilgili kutuphaneleri import edip kullanima hazir hale getirelim.
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset("diamonds")
df = diamonds.copy()
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
df.head()
df.cut.head()
```

```python
# cut kategorik degiskeni ile price sayisal degiskenini caprazlayarak bir goruntu olusturalım
```

```python
sns.catplot(x= "cut", y = "price", data=df);
```

```python
sns.barplot(x= "cut", y= "price", hue = "color", data=df);
```

#### Buradaki price degiskenleri iki groupby islemi sonrası iki kategorik degiskenin price ortalaması. Üzerindeki çubuklar da std. 
Bu edindiğimiz grafik için bir doğrulama işlemi yapalım.

```python
df.groupby(["cut","color"])["price"].mean()
```
